```python
urls = [
    "https://www.allocine.fr/membre-Z20041207222824320473123/films/",
    "https://www.allocine.fr/membre-Z20090118230847590126158/films/",
    "https://www.allocine.fr/membre-Z20060120185249557812865/films/",
    "https://www.allocine.fr/membre-Z20120318112307003102189/films/",
    "https://www.allocine.fr/membre-Z20110624203517947716343/films/",
    "https://www.allocine.fr/membre-Z20080928170535083698124/films/",
]
```

```python
url_max_page = []

for url in urls:
    session = requests.Session()
    url_text = session.get(url).text
    soup = BeautifulSoup(url_text, "html.parser")
    paragraphs = soup.find_all("nav", class_="pagination cf")
    if paragraphs:
        for paragraph in paragraphs:
            page_max = paragraph.select("[class*='button button-md item']")[
                -1
            ].text.strip()
            url_max_page.append((url, page_max))
    else:
        url_max_page.append((url, 0))

url_max_page
```

```python
all_ratings_movies = []

for item in url_max_page:
    session = requests.Session()
    url_text = session.get(item[0]).text
    soup = BeautifulSoup(url_text, "html.parser")
    cards = soup.find_all(
        "div", class_="card entity-card-simple userprofile-entity-card-simple"
    )
    for card in cards:
        dico = {}
        title_movie = card.find("a", class_="meta-title meta-title-link").text.strip()
        rating_movie = int(
            card.select("[class*='rating-mdl']")[0].get("class")[1].replace("n", "")
        )
        dico["movie"] = title_movie
        dico["rating_movie"] = rating_movie
        index_reviewer = url_max_page.index(item)
        all_ratings_movies.append((index_reviewer, dico))
    if int(item[1]) != 0:
        for page in range(2, int(item[1])):
            url = f"{item[0]}?page={page}"
            print(url)
            url_text = session.get(url).text
            soup = BeautifulSoup(url_text, "html.parser")
            cards = soup.find_all(
                "div", class_="card entity-card-simple userprofile-entity-card-simple"
            )
            for card in cards:
                dico = {}
                title_movie = card.select("[class*='meta-title meta-title-link']")[0].get("title")
                print(title_movie)
                rating_movie = int(
                    card.select("[class*='rating-mdl']")[0].get("class")[1].replace("n", "")
                )
                dico["movie"] = title_movie
                dico["rating_movie"] = rating_movie
                index_reviewer = url_max_page.index(item)
                all_ratings_movies.append((index_reviewer, dico))
```

```python
all_ratings_movies
```

```python
df_ratings = pd.DataFrame()
data_list = []

for item in all_ratings_movies:
    data_dict = {
        "userID": item[0],
        "movie": item[1]["movie"],
        "rating": item[1]["rating_movie"] / 100,
    }
    data_list.append(data_dict)

df_ratings = pd.DataFrame(data_list)
```

```python
def list_to_categorical_func(df, item):
    list_unique_values_col = df[item].unique().tolist()
    df[f"{item}ID"] = df[item].replace(
        list_unique_values_col,
        list(range(1, len(list_unique_values_col) + 1)),
        inplace=False,
    )
```
